### Checking for duplicates

```python
# checking for duplicates (how many duplicate rows are there in the dataset?)
df.duplicated().sum()
```

### The following code
#### returns all the duplicate rows in the dataset. Fro example, if the data is like this:<br>[1,1,2,3,3] it will return the rows with 1 and 3 because they are duplicated. <br>If we set keep = False, it will return all the duplicate rows including the first occurrence.<br>So in this case, it will return all the rows with 1 and 3. 

```python
df.loc[df.duplicated(keep= False)]
```

### Sorts duplicate rows by 'track_id'

```python
df.loc[df.duplicated(keep= False)].sort_values('track_id')
```

### To check for duplicates in a specific column, we can use the following code:

```python

df.loc[df['track_id'].duplicated(keep= False)].sort_values('track_id')
```

### multivariate (two or more variables) analysis

```python

df.columns
df_corr =df[['popularity',
       'duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr
```

### Heat map is a good tool to identify if there is  correlation between variables

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(df_corr, annot=True,cmap='mako', ax=ax)
#sns.heatmap(df_corr, ax=ax)
```
